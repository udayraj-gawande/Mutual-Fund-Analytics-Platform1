```python
import pandas as pd

df = pd.read_csv("../data/raw/02_nav_history.csv")

print(df.head())
print(df.info())
```

```python
print(df.dtypes)
```

```python
df = df.sort_values(by=['amfi_code', 'date'])
print(df.isnull().sum())
df['nav'] = df['nav'].ffill()
```

```python
print(df.duplicated().sum())
```

```python
print((df['nav'] <= 0).sum())

df = df[df['nav'] > 0]
```

```python
df.to_csv("../data/processed/clean_nav_history.csv", index=False)
```

```python
print(df.info())
print(df.isnull().sum())
print(df.describe())
```
